```python
import pandas as pd
corp_info = pd.read_csv('corp.csv', header=0, encoding = "CP949")
corp_info
```

```python
corp_info = corp_info[['종목코드', '종목명', '종가', '상장주식수']]
corp_info
```

```python
def addA(value):
    return 'A' + str(value)
corp_info['종목코드'].apply(addA)
corp_info
```

```python
corp_result = pd.read_excel('SAMPLE_2018.xlsx')
corp_result
```

```python
corp_result.columns
```

```python
corp_data = corp_result[['종목코드', '당기순이익']]
corp_data
```

```python
corp_total = pd.merge(corp_info, corp_data, how='left', on='종목코드')
corp_total
```

```python
corp_total = corp_info.join(corp_data.set_index('종목코드'), how='outer', on='종목코드')
corp_total
```
